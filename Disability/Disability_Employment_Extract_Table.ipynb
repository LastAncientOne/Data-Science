# Employment Disability Status Extract Table

```python
import pandas as pd
import urllib.request

from html_table_parser import HTMLTableParser
```

```python
web = 'https://www.bls.gov/news.release/disabl.t01.htm'
```

```python
req = urllib.request.Request(url=web)
f = urllib.request.urlopen(req)
xhtml = f.read().decode('utf-8')
```

```python
p = HTMLTableParser()
p.feed(xhtml)
```

```python
Table_List = p.tables
```

```python
Disability_stats = pd.DataFrame.from_dict(Table_List[0])
```

```python
Disability_stats 
```

```python
Disability_stats.to_excel("employment_disability.csv", na_rep='', float_format=None,                                columns=None, header=True, index=True, index_label=None,                                    startrow=0, engine=None, merge_cells=True, encoding=None,
                         inf_rep='inf', verbose=True)
```
