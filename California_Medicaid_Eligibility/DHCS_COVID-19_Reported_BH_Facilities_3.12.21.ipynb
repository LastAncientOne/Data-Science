```python
top_20 = df.groupby(['County'])['Staff: Confirmed COVID-19 (CUMULATIVE)'].sum().sort_values(ascending=False).head(20)
```

```python
top_20_R = df.groupby(['County'])['Residents/Clients: Confirmed COVID-19 (CUMULATIVE)'].sum().sort_values(ascending=False).head(20)
```

```python
colors =  ['b', 'g', 'r', 'c', 'm', 'y', 'k', 'pink', 'plum', 'coral']
plt.figure(figsize=(16,12))
plt.bar(top_20.keys(), top_20, color = colors)
plt.title('Staff: Confirmed COVID-19 (CUMULATIVE)')
plt.xticks(rotation=45, ha='right')
```

```python
colors =  ['b', 'g', 'r', 'c', 'm', 'y', 'k', 'pink', 'plum', 'coral']
plt.figure(figsize=(16,12))
plt.bar(top_20.keys(), top_20_R, color = colors)
plt.title('Residents/Clients: Confirmed COVID-19 (CUMULATIVE')
plt.xticks(rotation=45, ha='right')
```

```python
df['Facility Type'].value_counts().keys()
```

```python
x = df['Facility Type'].value_counts().values
y = df['Facility Type'].value_counts().keys()

explode = (0.2, 0.01, 0.1, 0.2, 0.4, 0.1)

fig1, ax1 = plt.subplots()
ax1.pie(x, explode=explode, labels=y, autopct='%1.1f%%',
        labeldistance=1.05)
ax1.axis('equal')  
plt.title('Total of Facility Type')
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(12,12))
plt.pie(x, labels=y, autopct='%1.1f%%', pctdistance=0.9, explode=(0,0,0,0,0.2,0.4))
plt.legend(loc='upper left', bbox_to_anchor=(-0.1, 1.))
plt.title('Total of Facility Type', fontsize=16)
plt.show()
```
