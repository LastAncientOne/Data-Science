# California Home Price

## Data Analysis

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('California_City_Homes_Prices_clean.csv')
```

```python
df.head()
```

```python
df.tail()
```

```python
df.info()
```

```python
df = df.set_index('Date')
```

```python
df.head()
```

```python
import matplotlib.pyplot as plt
plt.style.use('seaborn-darkgrid')
```

```python
df.plot(figsize=(20,16))
# Add legend
plt.legend(loc='center left', bbox_to_anchor=(1, 0.5))
 
# Add titles
plt.title("CA Home Price", loc='left', fontsize=16, fontweight=0, color='orange')
plt.xlabel("Date")
plt.ylabel("Price")

# Show the graph
plt.show()
```

```python
df.plot.hist(alpha=0.5, figsize=(20,16))
plt.legend(loc='center left', bbox_to_anchor=(1, 0.5))
```

```python
df.diff().hist(alpha=0.5, bins=50, figsize=(20, 16));
```

```python
df.plot.box(figsize=(20, 16))
plt.xticks(rotation=90)
plt.title("Box plots of California Cities", fontsize=16, fontweight=0, color='orange')
```

```python
color = {"boxes": "DarkGreen",
    "whiskers": "DarkOrange",
    "medians": "DarkBlue",
    "caps": "Gray"}

df.plot.box(color = color, figsize=(20, 16))
plt.xticks(rotation=90)
plt.title("Box plots of California Cities", fontsize=16, fontweight=0, color='orange')
```

```python
df.plot(subplots=True, layout=(12, 5), figsize=(20, 20), sharex=False);
```

```python
df.iloc[:, 0:10]
```

```python
df.iloc[56:61, :]
```

```python
from pandas.plotting import table

fig, ax = plt.subplots(figsize=(20,16))
df.iloc[:, 0:10].plot(ax=ax, ylim=(0, 1100000), legend=None)
table(ax,np.round(df.iloc[:, 0:10].describe(), 2), loc="bottom", bbox=[0.0,-0.35,1,.28])
plt.subplots_adjust(bottom=0.3)
plt.legend(loc='best')
plt.title('Top 10 Homes in CA',  fontsize=16, fontweight=0, color='orange')
```

```python
df.iloc[-1].sort_values(ascending=False)
```

```python
top_10 = df.iloc[-1].sort_values(ascending=False).head(10)
top_10
```

```python
top_10.values
```

```python
top_10.keys()
```

```python
# Convert series to dataframe
df_10 = top_10.to_frame()
df_10
```

```python
plt.figure(figsize=(16,10))
plt.bar(top_10.keys(), top_10.values, color = list('rgbkymc'))
plt.title('Top 10 Expensive Homes in CA',  fontsize=18, fontweight=0, color='orange')
```
