# Global Temperature Analysis

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_excel('Global_Temperature.xlsx')
```

```python
data.head()
```

```python
data.tail()
```

```python
data.describe()
```

```python
plt.plot(data['Year'], data['No_Smoothing'])
plt.title('No Smoothing')
plt.xlabel('Date')
plt.ylabel('Value')
```

```python
plt.plot(data['Year'], data['Lowess(5)'])
plt.title('Lowess')
plt.xlabel('Date')
plt.ylabel('Value')
```

```python
viz = data[['No_Smoothing','Lowess(5)']]
viz.hist()
plt.show()
```

```python
plt.style.use('seaborn-whitegrid')
fig = plt.figure()
ax = plt.axes()

plt.plot(data['Year'], data['No_Smoothing'])
plt.plot(data['Year'], data['Lowess(5)'])
plt.legend(loc='best')
plt.show()
```
