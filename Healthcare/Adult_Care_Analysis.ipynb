```python
bed = df.groupby(['County'])['Number of Beds'].sum().reset_index()
bed.sort_values('Number of Beds', ascending=False)
```

```python
plt.figure(figsize=(40,10))
plt.bar(bed['County'], bed['Number of Beds'])
plt.ylabel('Number of Beds')
plt.xlabel('Name of County')
plt.title('Total Number of Beds in each County')
```

```python
plt.figure(figsize=(40,10))
sns.set(style="whitegrid")
ax = sns.barplot(x="County", y="Number of Beds", data=bed)
```
