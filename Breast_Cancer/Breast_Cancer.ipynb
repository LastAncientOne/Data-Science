```python
import seaborn as sns

for i in range(0, len(dataset.columns), 5):
    sns.pairplot(data=dataset,
                x_vars=dataset.columns[i:i+5],
                y_vars=['target'])
```

```python
import operator

individual_features_df = []
for i in range(0, len(dataset.columns) - 1): # -1 because the last column is SalePrice
    tmpDf = dataset[[dataset.columns[i], 'target']]
    tmpDf = tmpDf[tmpDf[dataset.columns[i]] != 0]
    individual_features_df.append(tmpDf)

all_correlations = {feature.columns[0]: feature.corr()['target'][0] for feature in individual_features_df}
all_correlations = sorted(all_correlations.items(), key=operator.itemgetter(1))
for (key, value) in all_correlations:
    print("{:>15}: {:>15}".format(key, value))
```

```python
golden_features_list = [key for key, value in all_correlations if abs(value) >= 0.5]
print("There is {} strongly correlated values with Target:\n{}".format(len(golden_features_list), golden_features_list))
```

```python
corr = dataset.drop('target', axis=1).corr() # We already examined targetcorrelations
plt.figure(figsize=(12, 10))

sns.heatmap(corr[(corr >= 0.5) | (corr <= -0.4)], 
            cmap='viridis', vmax=1.0, vmin=-1.0, linewidths=0.1,
            annot=True, annot_kws={"size": 8}, square=True)
```

```python
# Seperate train and test data
from sklearn.model_selection import train_test_split

X = dataset.loc[:,dataset.columns!='target']
Y = dataset['target']
# Random State makes input fixed value
X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size=0.2, random_state=0)
```

```python
from sklearn.linear_model import LinearRegression
# 1. Set up the model
model = LinearRegression()
# 2. Use fit
model.fit(X_train, y_train)
# 3. Check the score
model.score(X_test, y_test)
```

```python
# Negative coefficient just means that as X increases, Y is predicted to decrease.
print('Coefficients:', model.coef_) # Get the coefficients
```

```python
print('intercept:', model.intercept_) # Get the intercept
```

```python
print('Predict:', model.predict(X_test)) # Predict data
```

In this case, breast cancer dataset is not good for linear regression because we have target is 0 or 1. Therefore, it means the patient has cancer or not. 

```python
from sklearn.linear_model import LogisticRegression

X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size=0.2, random_state=0)
```

```python
# 1. Set up the model
logreg = LogisticRegression(solver='liblinear')
# 2. Use fit
logreg.fit(X_train, y_train)
# 3. Check the score
logreg.score(X_test, y_test)
```

```python
print('Predict:', logreg.predict(X_test)) # Predict data
```

```python
y_pred = logreg.predict(X_test)
```

```python
from sklearn.metrics import confusion_matrix, classification_report, accuracy_score
#Print confusion matrix
print('Confusion matrix is as:')
print(confusion_matrix(y_test, y_pred))
```

```python
print(pd.DataFrame(confusion_matrix(y_test, y_pred), index = ['Actual 0', 'Actual 1'], columns = ['Predicted 0', 'Predicted 1' ]))
```

```python
print('Classification report is as:')
print(classification_report(y_test, y_pred))
acc = accuracy_score(y_test, y_pred)
print('Accuracy of model is {0}'.format(acc))
```

## Improve the model

```python
# Increase the test size and random state to 42
X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size=0.3, random_state=42)
```

```python
# 1. Set up the model
logreg = LogisticRegression(solver='liblinear')
# 2. Use fit
logreg.fit(X_train, y_train)
# 3. Check the score
logreg.score(X_test, y_test)
```

```python
y_pred = logreg.predict(X_test)
```

```python
print(pd.DataFrame(confusion_matrix(y_test, y_pred), index = ['Actual 0', 'Actual 1'], columns = ['Predicted 0', 'Predicted 1' ]))
```

```python
print('Classification report is as:')
print(classification_report(y_test, y_pred))
acc = accuracy_score(y_test, y_pred)
print('Accuracy of model is {0}'.format(acc))
```
