# Diabetes

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

from sklearn import datasets
```

```python
diabetes = datasets.load_diabetes()
```

```python
diabetes
```

```python
data = np.c_[diabetes.data, diabetes.target]
columns = np.append(diabetes.feature_names, ["target"])
dataset = pd.DataFrame(data, columns=columns)
```

```python
dataset.head()
```

```python
print("--------Exploring Diabetes Data --------")
print("Diabetes Data")
print('-'*60)
print("Dataset information") 
print(dataset.info(memory_usage='deep',verbose=False))
print('-'*60)
print(dataset.info())
print('-'*60)
print("Data type:")
print(dataset.dtypes)
print('-'*60)
print("Check unique values wihtout NaN")
print(dataset.nunique())
print('-'*60)
print("Data shape:")
print(dataset.shape)
print('-'*60)
print("Data columns Names:")
print(dataset.columns)
print('-'*60)
print("Check for NaNs:")
print(dataset.isnull().values.any())
print('-'*60)
print("Data Statistics Summary:")
print(dataset.describe())
# if have a NaN
print('-'*60)
print(dataset.isnull().sum())
```

```python
dataset.corr(method='pearson')
```

```python
dataset.corr(method='kendall')
```

```python
dataset.hist(sharex=False, sharey=False, xlabelsize=1, ylabelsize=1, figsize=(12,6))
plt.show()
```

```python
dataset.plot(kind='density', subplots=True, layout=(4,4), sharex=False, legend=False, fontsize=1, figsize=(12,6))
plt.show()
```

```python
dataset.plot(kind='box', subplots=True, layout=(4,4), sharex=False, sharey=False, fontsize=8, figsize=(12,10))
plt.show()
```

```python
from pandas.tools.plotting import scatter_matrix

scatter_matrix(dataset, figsize=(12,12))
plt.show()
```

```python
# Seperate train and test data
from sklearn.model_selection import train_test_split

X = dataset.loc[:,dataset.columns!='target']
Y = dataset['target']
# Random State makes input fixed value
X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size=0.2, random_state=0)
```

```python
from sklearn.linear_model import LinearRegression
# 1. Set up the model
model = LinearRegression()
# 2. Use fit
model.fit(X_train, y_train)
# 3. Check the score
model.score(X_test, y_test)
```

```python
# A positive sign indicates that as the predictor variable increases, the response variable also increases. 
# A negative sign indicates that as the predictor variable increases, the response variable decreases.
print('Coefficients:', model.coef_) # Get the coefficients, beta
```

```python
print('intercept:', model.intercept_) # Get the intercept
```

```python
print('Predict:', model.predict(X_test)) # Predict data
```

```python
# plot prediction and actual data
y_pred = model.predict(X_test) 
plt.plot(y_test, y_pred, '.')

# plot a line, a perfit predict would all fall on this line
x = np.linspace(0, 330, 100)
y = x
plt.plot(x, y)
plt.show()
```
